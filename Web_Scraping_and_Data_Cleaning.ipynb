```python
#Chenge column name from "'Critically_Rated_For_name'" to "Critically_Rated" and from 'Highly_Rated_For' to 'Highly Rated'
df.rename(columns={'Critically_Rated_For_name':'Critically_Rated','Highly_Rated_For':'Highly_Rated'},inplace=True)
df.columns
```

```python
#modify the two columns "Interviews"
def process_value(value):
    if 'k' in value:
        return float(value.replace('k', '')) * 1000
    else:
        return float(value)
df['Interviews'] = df['Interviews'].apply(process_value)
```

```python
#modify the two columns "Reviews"
def process_value(value):
    if 'k' in value:
        return float(value.replace('k', '')) * 1000
    else:
        return float(value)
df['Reviews'] = df['Reviews'].apply(process_value)
```

```python
#convert "Reviews" and "Interviews" to integers
df['Reviews'] = df['Reviews'].astype(int)
df['Interviews'] = df['Interviews'].astype(int)
df.head()
```

```python
#work on spliting the column industry/Loaction into two at the '|'
# Split the column into two new columns
df[['Industry', 'Locations']] = df['Industry/Locations'].str.split('|', expand=True)
```

```python
# Define new column order
new_columns = ['Name', 'Industry', 'Locations', 'Highly_Rated', 'Critically_Rated', 'Rating', 'Reviews', 'Interviews']
```

```python
# Reorder columns
df = df[new_columns]
```

```python
# Print new column order
print("\nNew Column Order:")
print(df.columns.tolist())
```

```python
df['Reviews'] = df['Reviews'] / 1000
df['Interviews'] = df['Interviews'] / 1000
df.head()
```

```python
df['Locations'].head()
```

```python
#work on the locations column to come up with another column that displays the exact number of locations each compnay has
# Function to extract the number of locations
def extract_location_count(value):
  parts = value.split('+')# Split the string at the '+' sign and take the second part
  if len(parts) > 1:
    number = ''.join(filter(str.isdigit, parts[1]))
    return int(number) + 1 #add 1 (since the main location is also counted)
  else:
    return 1 #If there's no '+' sign, assume it's just one location
```

```python
# Apply the function to the 'Locations' column
df['Number_of_Locations'] = df['Locations'].apply(extract_location_count)
```

Exploratory Data Analysis


```python
#Count the Number of different idustires in the data
df['Industry'].value_counts()
```

```python
#Calculate the avarage rating per Industry
industry_rating = df.groupby('Industry')['Rating'].mean()
industry_rating
```

```python
#the rating should be in two decimal places
industry_rating = industry_rating.round(2)
industry_rating
```

```python
#plot the data
plt.figure(figsize=(10, 6))
industry_rating.plot(kind='bar')
plt.xlabel('Industry')
plt.ylabel('Average Rating')
plt.title('Average Rating by Industry')
plt.show()
```

```python
#Check the Number of interviews per Company
company_interviews = df.groupby('Name')['Interviews'].mean()
company_interviews
```

```python
#plot a bar graph it should be big
plt.figure(figsize=(15, 6))
company_interviews.plot(kind='bar')
plt.xlabel('Company')
plt.ylabel('Average Interviews')
plt.title('Average Interviews by Company')
plt.show()
```

```python
#pivot table for Average reviews and interviews by company
pivot_table = df.pivot_table(index='Name', values=['Reviews', 'Interviews'], aggfunc='mean')
pivot_table
```

```python
# Bar Chart for Interviews vs. Reviews
plt.figure(figsize=(15, 6))
plt.bar(df['Name'], df['Interviews'], label='Interviews')
plt.bar(df['Name'], df['Reviews'], bottom=df['Interviews'], label='Reviews')
plt.title('Interviews and Reviews by Company')
plt.xlabel('Company Name')
plt.ylabel('Number')
plt.legend()
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Bar Chart for Interviews and Line Graph for Reviews
plt.figure(figsize=(15, 6))
plt.bar(df['Name'], df['Interviews'], label='Interviews')
plt.plot(df['Name'], df['Reviews'], marker='o', linestyle='-', color='red', label='Reviews')
plt.title('Interviews and Reviews by Company')
plt.xlabel('Company Name')
plt.ylabel('Number')
plt.legend()
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
